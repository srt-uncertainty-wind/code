```python
# 
import numpy as np
import pandas as pd
import os
import plotly.graph_objects as go

# 
simu_folder_path = '../Dataset/Simulate_Data/set_1'
data_excel_path = os.path.join(simu_folder_path, 'File.xlsx')
para_info_txt_path = os.path.join(simu_folder_path, 'para_info.txt')

# nasa
variable_list = ['ALT', 'ALTR', "TAS", 'GS', 'AOA1', 'AOA2', 'PTCH', 'WS', "WD", 'SAT', 'TAT', 'PI', 'PT', 'GAMMA']

# 
simu_record_list = pd.read_excel(data_excel_path).values[:, 0]
simu_array = np.array([list(filter(None, record.split(' '))) for record in simu_record_list]).astype('float32')
print(simu_array.shape)

with open(para_info_txt_path, 'r') as f:
    simu_variable_list = f.readline().strip().split(', ')
print(simu_variable_list)

# nasa
nasa_simu_order_dict = {0: 1, 1: 2, 2: 3, 4: 4, 13: 5, 6: 6}
simu_variable_list_ordered = [simu_variable_list[nasa_simu_order_dict[idx]] if idx in nasa_simu_order_dict.keys() else 'None' for idx in range(len(variable_list))]
simu_array_ordered = np.array([simu_array[:, nasa_simu_order_dict[idx]] if idx in nasa_simu_order_dict.keys() else [None for _ in range(simu_array.shape[0])] for idx in range(len(variable_list))]).T
print(simu_variable_list_ordered)
print(simu_array_ordered.shape)

# 
for idx in range(len(variable_list)):
    if simu_variable_list_ordered[idx] != 'None':
        fig = go.Figure()
        # fig.add_trace(go.Scatter(y=test_X_origin[:decay_steps+pred_steps, idx], mode='lines', name = 'GT', line = dict(color='blue')))
        fig.add_trace(go.Scatter(y=simu_array_ordered[:, idx], mode='lines', line = dict(color='blue')))
        fig.update_layout(
            title=f'Plot of {simu_variable_list_ordered[idx]}',
            xaxis_title='Time Step',
            yaxis_title='Value'
        )
        fig.show()
```
