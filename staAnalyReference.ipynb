```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import os

# 1. 加载数据
root_path = "../Data_Summary"
filename = os.path.join(root_path, "DASHlink_full_fourclass_raw_comp.npz")
full_data = np.load(filename)
data = full_data['data']
label = full_data['label']

variables = ['AILERON POSITION LH', 'AILERON POSITION RH', 'CORRECTED ANGLE OF ATTACK', 'BARO CORRECT ALTITUDE LSP', 
             'COMPUTED AIRSPEED LSP', 'SELECTED COURSE', 'DRIFT ANGLE', 'ELEVATOR POSITION LEFT', 'T.E. FLAP POSITION', 
             'GLIDESLOPE DEVIATION', 'SELECTED HEADING', 'LOCALIZER DEVIATION', 'CORE SPEED AVG', 'TOTAL PRESSURE LSP', 
             'PITCH ANGLE LSP', 'ROLL ANGLE LSP', 'RUDDER POSITION', 'TRUE HEADING LSP', 'VERTICAL ACCELERATION', 'WIND SPEED']
```

```python
# 2. 绘制每个变量随时间的变化图
timeVar_save_path = os.path.join(root_path, "figures", "timeVar")
if os.path.exists(timeVar_save_path) == False:
    os.mkdir(timeVar_save_path)
for variable_index, variable_name in enumerate(variables):
    plt.figure(figsize=(10, 5))
    plt.plot(data[0, :, variable_index])  # 绘制第一个实例的该变量随时间的变化
    plt.title(variable_name + ' over time')
    plt.savefig(os.path.join(timeVar_save_path, variable_name + '_over_time.png'))
    plt.show()
```

```python
# 3. 根据类别绘制数据分布的直方图
plt.hist(label, bins=4, range=(0,4), align='left')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Nominal', 'Speed High', 'Path High', 'Flaps Late Setting'])
plt.title('Distribution of Classes')
plt.savefig(os.path.join(root_path, "figures", "class_distribution.png"))
plt.show()
```

```python
# 4. 数据预处理
# Z-Score Normalization
scaler_zscore = StandardScaler()
scaled_data_zscore = scaler_zscore.fit_transform(data.reshape(-1, 20)).reshape(data.shape)

# Min-Max Normalization
scaler_minmax = MinMaxScaler()
scaled_data_minmax = scaler_minmax.fit_transform(data.reshape(-1, 20)).reshape(data.shape)
```

```python
# 5. 可视化与描述性统计
stas_path = os.path.join(root_path, "figures", "statistics")
if os.path.exists(stas_path) == False:
    os.mkdir(stas_path)
for variable_index, variable_name in enumerate(variables):
    variable_stas_path = os.path.join(stas_path, variable_name)
    if os.path.exists(variable_stas_path) == False:
        os.mkdir(variable_stas_path)
    # 可视化
    plt.figure(figsize=(15, 5))
    
    # 原始数据
    plt.subplot(1, 3, 1)
    plt.hist(data[:, :, variable_index].flatten(), bins=50, color='blue', alpha=0.7)
    plt.title(f'Original Data for {variable_name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    # Z-Score标准化数据
    plt.subplot(1, 3, 2)
    plt.hist(scaled_data_zscore[:, :, variable_index].flatten(), bins=50, color='green', alpha=0.7)
    plt.title(f'Z-Score Normalized for {variable_name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    # Min-Max归一化数据
    plt.subplot(1, 3, 3)
    plt.hist(scaled_data_minmax[:, :, variable_index].flatten(), bins=50, color='orange', alpha=0.7)
    plt.title(f'Min-Max Normalized for {variable_name}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.savefig(os.path.join(variable_stas_path, f'{variable_name}.png'))
    plt.show()

    # 描述性统计
    print(f"Original Data for {variable_name}: Mean = {data[:, :, variable_index].flatten().mean()}, Std Dev = {data[:, :, variable_index].flatten().std()}")
    print(f"Z-Score Normalized Data for {variable_name}: Mean = {scaled_data_zscore[:, :, variable_index].flatten().mean()}, Std Dev = {scaled_data_zscore[:, :, variable_index].flatten().std()}")
    print(f"Min-Max Normalized Data for {variable_name}: Mean = {scaled_data_minmax[:, :, variable_index].flatten().mean()}, Std Dev = {scaled_data_minmax[:, :, variable_index].flatten().std()}")
    print("---------------------------------------------------------")
```
